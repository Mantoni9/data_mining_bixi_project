```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
stations_map_df = pd.read_csv('data/stations_preprocessed/Stations_2019.csv')
stations_map_df.head()
```

```python
stations_map_df.head()
```

```python
BBox = (stations_map_df.longitude.min(), stations_map_df.longitude.max(), stations_map_df.latitude.min(), stations_map_df.latitude.max())
print(BBox)
```

```python
montreal_map = plt.imread('data/map_of_montreal.png')
```

```python
fig, ax = plt.subplots(figsize = (15,15))
ax.scatter(stations_map_df.longitude, stations_map_df.latitude, zorder=1, alpha= 0.2, c='b', s=20)
ax.set_title('Plotting bixi bike station data')
ax.set_xlim(BBox[0],BBox[1])
ax.set_ylim(BBox[2],BBox[3])
ax.imshow(montreal_map, zorder=0, extent = BBox, aspect= 'equal')
```
